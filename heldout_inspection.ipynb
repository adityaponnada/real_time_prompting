```python
def compute_heldout_first_fraction_completion(heldout_df, participant_col='participant_id', outcome_col='outcome', fraction=0.1, sort_col=None):
    """Compute the first `fraction` of observations per participant and per-participant completion rates."""
    import pandas as pd
    import numpy as np
    if heldout_df is None or participant_col not in heldout_df.columns:
        raise ValueError('heldout_df must contain the participant column')
    if outcome_col not in heldout_df.columns:
        raise ValueError('outcome_col must be a column in heldout_df')
    if not (0 <= fraction <= 1):
        raise ValueError('fraction must be between 0 and 1')
    parts = []
    rows = []
    for pid, g in heldout_df.groupby(participant_col):
        if sort_col is not None and sort_col in g.columns:
            g = g.sort_values(sort_col)
        else:
            g = g.sort_index()
        n = len(g)
        if n == 0:
            continue
        k = int(np.floor(n * fraction))
        if fraction > 0 and k == 0:
            k = 1
        sel = g.iloc[:k]
        parts.append(sel)
        n_selected = len(sel)
        n_completed = int(sel[outcome_col].eq(1).sum())
        completion_rate = n_completed / n_selected if n_selected > 0 else np.nan
        rows.append({participant_col: pid, 'n_selected': n_selected, 'n_completed': n_completed, 'completion_rate': completion_rate})
    subset_df = pd.concat(parts, ignore_index=True) if parts else pd.DataFrame(columns=heldout_df.columns)
    rates_df = pd.DataFrame(rows, columns=[participant_col, 'n_selected', 'n_completed', 'completion_rate'])
    return subset_df, rates_df

# Example usage:

subset_df, rates_df = compute_heldout_first_fraction_completion(heldout_df, participant_col='participant_id', outcome_col='outcome', fraction=0.1, sort_col=None)# Adjust `outcome_col` if your outcome column has a different name (e.g. 'completed')
```
